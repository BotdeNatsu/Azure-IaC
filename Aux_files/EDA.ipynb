```python
sd.seedir(
    IMG_TRAIN_PATH, 
    style='emoji', 
    indent=4, 
    depthlimit=2, 
    beyond='content'
)
```

```python
def img_from_plate(filepath, num_plate):
   paths = []
   for root, dirs, files in os.walk(filepath):
      for file in files:
        if num_plate and (num_plate in file):
            paths.append(os.path.join(root, file))
   return(paths)

IMG_PATHS = img_from_plate(IMG_TRAIN_PATH, '270389')
for img_path in IMG_PATHS:
   img = plt.imread(img_path)
   plt.imshow(img)
   plt.title(img_path)
   plt.show()
    
```
